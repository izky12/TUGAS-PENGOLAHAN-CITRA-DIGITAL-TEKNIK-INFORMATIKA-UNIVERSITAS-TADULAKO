```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Backup bussid/mammrogram.jpg", 0)

img_1 = 255 - img


cv2.putText(img, "Original Image", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
cv2.putText(img_1, "Image Negatif", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

cv2_imshow(img)
cv2_imshow(img_1)
```
